```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars=pd.read_csv("cars-dataset.csv")
cars.head()
```

```python
cars.isna().sum()
```

```python
cars.drop("Serie",axis=1)
```

```python
cols=["Model","Company","Cylinders","Displacement","Power(HP)","Torque(lb-ft)","Fuel","Top Speed","Acceleration 0-62 Mph (0-100 kph)","Specification summary"]
```

```python
cars.iloc[:,0:21]
```

```python
df=cars[cols]
df.head()
```

```python
df.isna().sum()
```

```python
df[["Engine Size","Remaining"]]=df["Specification summary"].str.split(" ",n=1,expand=True)
df.drop(columns=["Engine Size"],inplace=True,axis=1)
df.drop(columns=["Remaining"],inplace=True,axis=1)
df.head()
```

```python
df["Displacement"]=df["Displacement"].str.replace("cm3","")
df.head()
```

```python
df.isna().sum()
df=df.dropna()
```

```python
df["Displacement"]=df["Displacement"].str.strip()
df["Displacement"]=df["Displacement"].str.replace(",","")
df["Displacement"].astype(float).astype(int)
```

```python
df["Displacement"] = (pd.to_numeric(df["Displacement"], errors='coerce') / 1000).round(1)
```

```python
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
df["Power(HP)"]=df["Power(HP)"].str.replace("HP","")
df["Power(HP)"]=df["Power(HP)"].str.replace("@","")
df["Power(HP)"]=df["Power(HP)"].str.replace("RPM","")
```

```python
df.head()
```

```python
df["Power(HP)"] = df["Power(HP)"].str.split(" ",expand=True)[0]
df.head()
```

```python
df["Torque(lb-ft)"]=df["Torque(lb-ft)"].str.replace("lb-ft","")
df["Torque(lb-ft)"]=df["Torque(lb-ft)"].str.replace("@","")
df["Torque(lb-ft)"]=df["Torque(lb-ft)"].str.replace("RPM","")
df.head()
```

```python
df["Top Speed"]=df["Top Speed"].str.replace("mph","")
df["Top Speed"]=df["Top Speed"].str.replace("km/h","")
df["Top Speed"]=df["Top Speed"].str.replace("(","")
df["Top Speed"]=df["Top Speed"].str.replace(")","")
df.head()
```

```python
df.head()
```

```python
df.head()
```

```python
df["Torque(lb-ft)"] = df["Torque(lb-ft)"].str.split(" ",expand=True)[0]
```

```python
df.head()
```

```python
df["Acceleration 0-62 Mph (0-100 kph)"]=df["Acceleration 0-62 Mph (0-100 kph)"].str.replace("s","")
```

```python
df.head()
```

```python
df.to_csv("cars_Data_Cleaned.csv")
```
